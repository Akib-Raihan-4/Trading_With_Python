## Price Return:
- Price return simply looks at how much the prices of the stocks in your portfolio have changed.
- For example, if you buy a stock for $100 and its price increases to $110, your price return is 10%.

$$ R_{t,t+1} = \frac{P_{t+1}- P_{t}}{P_{t}} $$

## Total Return:

- Total return considers not only the price changes of the stocks but also any additional income generated by the portfolio, such as dividends.
- Dividends are payments made by some companies to their shareholders. So, if you own a stock that pays dividends, you receive some extra money.

## Multi Period Returns

$$ R_{t,t+2} = (1 + R_{t,t+1})(1 + R_{t,t+2}) -1 $$

## Annualizing Returns

$$ R_{t,t+n} = (1+R_{t,t+n})^n -1 $$

**Question**: The return over the month is 1% what is the annualized return ? <br>
**Answer**: $$ ((1+0.01)^{12} - 1) $$

**Question** : The stock gains 1% over a quater (i.e a 3 month period). What is it's annualized return ?
**Answer** : $$ ((1+0.01)^4 - 1) * 100  = 4.06 $$
